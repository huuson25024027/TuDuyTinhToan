### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def count_k(n):
    dem = 0  # Tạo một biến đếm, ban đầu chưa có số nào thỏa mãn nên bằng 0
    moc_so_sanh = (n - 1) / 2     
    # Dùng vòng lặp for để duyệt các số nguyên dương k từ 1 đến sát n
    # Hàm range(1, n) sẽ tự động tạo ra các số 1, 2, 3... đến n-1
    for k in range(1, n):
        if k > moc_so_sanh:
            dem = dem + 1  # Nếu đúng, tăng biến đếm lên 1
    return dem
n = int(input("Nhập giá trị n: "))
ket_qua = count_k(n)
print(f"Số lượng các số k thỏa mãn là: {ket_qua}")
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
import math  
def triangle_area(a, b, c):
    if (a + b <= c) or (a + c <= b) or (b + c <= a):
        # Nếu vô lý, "ném" ra lỗi ValueError và dừng hàm ngay lập tức
        raise ValueError("Ba số này không thể tạo thành một tam giác!")
    p = (a + b + c) / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))    
    return dien_tich
a,b,c = map(int, input("Nhập 3 cạnh của tam giác: ").split())
ket_qua = triangle_area(a,b,c)
print(f"Diện tích tam giác hợp lệ là: {ket_qua}")
```

### Problem 7

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def shift_character(x, n):
    if x.islower():
        moc_bat_dau = ord('a') # số 97
    elif x.isupper():
        moc_bat_dau = ord('A') # số 65
    else:
        # Nếu nhập vào không phải là chữ cái (VD: số, dấu chấm...), trả về y nguyên
        return x       
    # Quy đổi chữ cái về vị trí từ 0 đến 25
    # Ví dụ: x là 'c' (99), mốc là 'a' (97) -> Vị trí là 99 - 97 = 2
    vi_tri_hien_tai = ord(x) - moc_bat_dau    
    # Tiến lên n bước và dùng % 26 để xoay vòng nếu lố qua chữ Z
    vi_tri_moi = (vi_tri_hien_tai + n) % 26   
    # Cộng trả lại điểm neo ban đầu để ra mã ASCII thật, rồi ép lại thành chữ
    ky_tu_moi = chr(vi_tri_moi + moc_bat_dau)  
    return ky_tu_moi
print(shift_character(x,n)) 
x = input("Nhập một ký tự: ")
n = int(input("Nhập số bước dịch chuyển: "))
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def to_lowercase(char):
    # Kiểm tra xem người dùng có nhập đúng 1 ký tự không
    if len(char) != 1:
        raise ValueError("Lỗi: Bạn phải nhập chính xác MỘT ký tự!")   
    # Nếu hợp lệ, chuyển thành chữ thường và trả kết quả ra
    return char.lower()
ky_tu_nhap = input("Nhập vào một ký tự in hoa: ")
ket_qua = to_lowercase(ky_tu_nhap)
print(f"Ký tự {ky_tu_nhap} sau khi chuyển thành chữ thường là: '{ket_qua}'")
```

### Problem 10
#### Write a function that, given a lowercase letter, returns a string containing that letter repeated 10 times.

```python
def repeat_letter_an_toan(char):
    if len(char) != 1 or not char.islower():
        raise ValueError("Lỗi: Bạn bắt buộc phải nhập MỘT chữ cái in thường (a-z)!")
    return char * 10
ky_tu = input("Nhập vào một chữ cái in thường: ")
ket_qua = repeat_letter_an_toan(ky_tu)
print(f"Chuỗi sau khi lặp 10 lần là: {ket_qua}")
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
def draw_triangle(n):
    # Tạo một danh sách rỗng để chứa các dòng của tam giác
    cac_dong = []  
    # Vòng lặp chạy từ 1 đến n (nhớ dùng n + 1 vì hàm range không lấy số cuối)
    for i in range(1, n + 1):
        # Dòng thứ i sẽ có i dấu sao. Đưa dòng này vào danh sách
        cac_dong.append("*" * i)       
    # Nối tất cả các dòng trong danh sách lại với nhau, cách nhau bằng dấu xuống dòng '\n'
    return "\n".join(cac_dong)
chieucao = int(input("Nhập chiều cao của tam giác (n): "))
ket_qua = draw_triangle(chieucao)
print(ket_qua)
```
