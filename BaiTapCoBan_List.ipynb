### Bài 1: Tìm số lớn nhất của 1 list

```python
lst = list(map(int, input("Nhập một dãy số nguyên: ").split()))
if lst:
    ans = lst[0]
    for i in lst:
        if i > ans:
            ans = i
    print("Số lớn nhất trong dãy là:", ans)
else:
    print("Bạn chưa nhập số nào.")
```

### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 
0 1

```python
def min_and_pos(lst):
    if not lst:
        return None, None  # Trả về None nếu danh sách rỗng
    min_value = lst[0]
    min_pos = 0
    for i in range(1, len(lst)):
        if lst[i] < min_value:
            min_value = lst[i]
            min_pos = i
    return min_value, min_pos

assert min_and_pos([3, 1, 4, 1, 5]) == (1, 1)
assert min_and_pos([]) == (None, None)
#assert để kiểm tra tính đúng đắn của hàm min_and_pos với các trường hợp khác nhau. Nếu tất cả assert đều đúng, chương trình sẽ chạy mà không có lỗi nào. Nếu có assert nào sai, chương trình sẽ dừng lại và báo lỗi, giúp bạn xác định vấn đề trong hàm min_and_pos.
```

### Bài 3: Tìm giá trị trung bình của 1 list

```python
lst = list(map(int, input("Nhập một dãy số nguyên: ").split()))
tong = 0
for i in lst:
    tong += i
gia_tri_trung_binh = tong / len(lst) if lst else 0
print("Giá trị trung bình của dãy số là:", {gia_tri_trung_binh})
```

### Bài 3b: Tìm giá trị trung bình của 1 list, nhưng chỉ tính những phần tử lớn hơn 0 (dương). Nếu không có số dương nào thì trả về invalid


```python
lst = list(map(int, input("Nhập một dãy số nguyên: ").split()))
if lst:
    lst_duong = [i for i in lst if i > 0]
    if lst_duong:
        for i in lst_duong:
            tong += i
        gia_tri_trung_binh_duong = tong / len(lst_duong)
        print("Giá trị trung bình của các số dương là:", gia_tri_trung_binh_duong)
    else:
        print("invalid")
else:
    print("Bạn chưa nhập số nào.")
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không

```python
def kiem_tra_doi_xung(lst):
    # lst[::-1] sẽ tạo ra một bản sao của list nhưng theo thứ tự đảo ngược
    if lst == lst[::-1]:
        return True
    return False

lst = list(map(int, input().split()))
print(f"List {lst} có đối xứng không? {kiem_tra_doi_xung(lst)}")
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89 => 
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def dua_zero_ve_dau(lst):
    # Lấy ra tất cả các số 0
    zeros = [x for x in lst if x == 0]
    # Lấy ra tất cả các số khác 0 (thứ tự tự động được giữ nguyên)
    non_zeros = [x for x in lst if x != 0]
    
    # Nối 2 list lại với nhau
    return zeros + non_zeros

test_list = [3, 4, 6, 7, 1, 9, 0, 0, 23, 0, 0, 34, 67, 89]
print("List ban đầu:", test_list)
print("List sau khi xử lý:", dua_zero_ve_dau(test_list))
```

### Bài 6: Xóa tất cả các số giống nhau trong list, Các số khác giữ nguyên vị trí trước sau

```python
def xoa_trung_lap_co_ban(lst):
    ket_qua = []  
    
    for so in lst:
        if so not in ket_qua:
            ket_qua.append(so)
            
    return ket_qua

chuoi_nhap_vao = input("Hãy nhập các số nguyên: ")
danh_sach_chu_so = chuoi_nhap_vao.split()

# Dùng vòng lặp for cơ bản để ép kiểu từng chữ số sang số nguyên
list_so_nguyen = []
for x in danh_sach_chu_so:
    list_so_nguyen.append(int(x))

print(f"\nList ban đầu bạn vừa nhập là: {list_so_nguyen}")

ket_qua_cuoi = xoa_trung_lap_co_ban(list_so_nguyen)
print(f"List sau khi xóa các số trùng lặp: {ket_qua_cuoi}")
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang
9 8 7 6 5 4 0: Giam
1 2 3 7 1 0 2: Vo danh
1 2 2 4 5 7: Vo danh

```python
def kiem_tra_day_so(lst):
    # Nếu list có ít hơn 2 phần tử, không đủ để kết luận xu hướng
    if len(lst) < 2:
        return "Vô danh"
        
    # Kiểm tra xem tất cả phần tử đứng trước có nhỏ hơn phần tử đứng liền sau không
    is_tang = all(lst[i] < lst[i+1] for i in range(len(lst) - 1))
    
    # Kiểm tra xem tất cả phần tử đứng trước có lớn hơn phần tử đứng liền sau không
    is_giam = all(lst[i] > lst[i+1] for i in range(len(lst) - 1))
    
    if is_tang:
        return "Tăng"
    elif is_giam:
        return "Giảm"
    else:
        return "Vô danh"

test_cases = [
    [1, 3, 5, 6, 7, 8, 9],
    [9, 8, 7, 6, 5, 4, 0],
    [1, 2, 3, 7, 1, 0, 2],
    [1, 2, 2, 4, 5, 7]
]

for tc in test_cases:
    print(f"{tc}: {kiem_tra_day_so(tc)}")
```

### Bài 8: Kiểm tra một list số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac

```python
def kiem_tra_ziczac(lst):
    if len(lst) < 2:
        return False
        
    if len(lst) == 2:
        if lst[0] != lst[1]:
            return True
        else:
            return False

    for i in range(1, len(lst) - 1):
        la_dinh = (lst[i] > lst[i-1]) and (lst[i] > lst[i+1])
        la_day = (lst[i] < lst[i-1]) and (lst[i] < lst[i+1])
        
        if not (la_dinh or la_day):
            return False
            
    return True

cac_test_case = [
    [1, 3, 2, 5, 2, 8, 7, 9, 2, 100],
    [34, 2, 56, 7, 90, 20, 30, 7, 120]
]

# Dùng vòng lặp duyệt qua từng test case và in kết quả
for tc in cac_test_case:
    print(f"{tc}: ", end="") 
    
    if kiem_tra_ziczac(tc):
        print("ziczac")
    else:
        print("khong phai ziczac")
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về "-1"

```python
def tim_vi_tri(lst, k):
    ket_qua = []  

    for i in range(len(lst)):
        if lst[i] == k:
            ket_qua.append(i)
            
    if len(ket_qua) == 0:
        return "-1"
    else:
        return ket_qua

danh_sach = [1, 3, 4, 4, 3, 3, 4, 3, 6, 1]

# Test case 1: k = 3
k1 = 3
print(f"k = {k1}, {danh_sach} => k xuất hiện tại các vị trí: {tim_vi_tri(danh_sach, k1)}")

# Test case 2: k = 10
k2 = 10
print(f"k = {k2}, {danh_sach} => k không xuất hiện trong ds => trả về: {tim_vi_tri(danh_sach, k2)}")
```

### Bài 10: 
Nhập vào tên một người, in ra kí tự đầu tiên và kí tự cuối cùng của tên đó.
VD: "Nguyen Ai Quoc" => ["N", "c"]

```python
def lay_ki_tu_dau_cuoi(ten):
    ki_tu_dau = ten[0]
    ki_tu_cuoi = ten[-1]
    return [ki_tu_dau, ki_tu_cuoi]

ten_nhap = input("Nhập vào tên một người: ")
print(f"Kết quả: {lay_ki_tu_dau_cuoi(ten_nhap)}")
```
