### W6A1 Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
chuoi_nhap = input("Nhập các số nguyên cách nhau bởi dấu cách: ")

# Tách chuỗi và chuyển thành list các số nguyên
danh_sach = []
for x in chuoi_nhap.split():
    danh_sach.append(int(x))

ket_qua = []
for so in danh_sach:
    if so not in ket_qua:  # Kiểm tra nếu số chưa có trong list kết quả thì mới thêm vào
        ket_qua.append(so)

print("List kết quả:", ket_qua)
```

### W6A2 Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị trí i), giữ nguyên thứ tự

```python
chuoi_nhap = input("Nhập các số nguyên cách nhau bởi dấu cách: ")
danh_sach = [int(x) for x in chuoi_nhap.split()] # Cách viết gọn để tạo list

tong_tich_luy = []
tong_hien_tai = 0

for so in danh_sach:
    tong_hien_tai += so
    tong_tich_luy.append(tong_hien_tai)

print("List tổng tích lũy:", tong_tich_luy)
```


### W6A3 Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp xếp.)


```python
nhap_tuple = input("Nhập các số nguyên cho tuple (cách nhau bởi dấu cách): ")
# Chuyển dữ liệu nhập vào thành kiểu tuple
tup = tuple(int(x) for x in nhap_tuple.split())

k = int(input("Nhập số nguyên k (số vị trí cần xoay): "))
n = len(tup)

if n > 0:
    k = k % n  # Đảm bảo k không vượt quá độ dài của tuple
    # Cắt tuple thành 2 phần và ghép lại: phần từ k đến cuối + phần từ đầu đến k
    tup_xoay = tup[k:] + tup[:k]
else:
    tup_xoay = tup

print("Tuple sau khi xoay trái:", tup_xoay)
```

### W6A4 Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp lại, thêm vào list của key đó

```python
nhap = input("Nhập các cặp key:value cách nhau bởi dấu cách (VD: a:1 b:2 a:3): ")
cac_cap = nhap.split()

tu_dien = {}
for cap in cac_cap:
    # Tách key và value dựa vào dấu hai chấm
    key, value = cap.split(':')
    
    if key not in tu_dien:
        tu_dien[key] = []  # Nếu key chưa tồn tại, tạo một list rỗng cho key đó
    tu_dien[key].append(value)  # Thêm value vào list của key

print("Dictionary thu được:", tu_dien)
```

### W6A5 Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ...,'zeros': ...} (đếm số >0, <0, =0)

```python
chuoi_nhap = input("Nhập danh sách các số nguyên cách nhau bởi dấu cách: ")
danh_sach = [int(x) for x in chuoi_nhap.split()]

dem = {'positives': 0, 'negatives': 0, 'zeros': 0}

for so in danh_sach:
    if so > 0:
        dem['positives'] += 1
    elif so < 0:
        dem['negatives'] += 1
    else:
        dem['zeros'] += 1

print("Kết quả đếm:", dem)
```

### W6A7 Viết chương trình nhập vào một tuple, in ra phần tử đầu, phần tử cuối, và tuple đảo ngược

```python
nhap = input("Nhập các phần tử của tuple cách nhau bởi dấu cách: ")
tup = tuple(nhap.split())

if len(tup) > 0:
    phan_tu_dau = tup[0]
    phan_tu_cuoi = tup[-1]
    # Dùng cú pháp [::-1] để đảo ngược list/tuple/chuỗi
    tup_dao_nguoc = tup[::-1]
    
    print("Phần tử đầu:", phan_tu_dau)
    print("Phần tử cuối:", phan_tu_cuoi)
    print("Tuple đảo ngược:", tup_dao_nguoc)
else:
    print("Tuple rỗng!")
```

### W6A9 Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của key.

```python
# Yêu cầu value phải là số nguyên để có thể cộng dồn
nhap1 = input("Nhập dict 1 (các cặp key:value cách nhau khoảng trắng, VD: a:1 b:2): ")
nhap2 = input("Nhập dict 2 (các cặp key:value cách nhau khoảng trắng, VD: b:3 c:4): ")

dict1 = {}
if nhap1:
    for cap in nhap1.split():
        k, v = cap.split(':')
        dict1[k] = int(v)
        
dict2 = {}
if nhap2:
    for cap in nhap2.split():
        k, v = cap.split(':')
        dict2[k] = int(v)

# Bắt đầu gộp
dict_tong = dict1.copy()
for k, v in dict2.items():
    if k in dict_tong:
        dict_tong[k] += v  # Cộng dồn nếu key đã có
    else:
        dict_tong[k] = v   # Thêm mới nếu key chưa có

# Sắp xếp dictionary theo key tăng dần
dict_ket_qua = {}
# Hàm sorted() mặc định sắp xếp các key theo thứ tự từ điển chữ cái (a -> z)
for k in sorted(dict_tong.keys()):
    dict_ket_qua[k] = dict_tong[k]

print("Dictionary mới sau khi gộp và sắp xếp:", dict_ket_qua)
```

### W6A10 Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.

```python
chuoi_nhap = input("Nhập danh sách các số nguyên (cách nhau bởi dấu cách): ")
danh_sach = [int(x) for x in chuoi_nhap.split()]
k = int(input("Nhập số nguyên dương k: "))

cap_so = set()
n = len(danh_sach)

# Dùng 2 vòng lặp lồng nhau để duyệt qua mọi cặp số
for i in range(n):
    for j in range(i + 1, n):
        if danh_sach[i] + danh_sach[j] == k:
            # Lưu thành tuple, giữ nguyên thứ tự xuất hiện (số thứ nhất đứng trước số thứ hai)
            cap = (danh_sach[i], danh_sach[j])
            cap_so.add(cap)

# Chuyển set thành list và sắp xếp theo phần tử đầu tiên của tuple
# lambda x: x[0] nghĩa là lấy tiêu chí sắp xếp là phần tử ở vị trí index 0 của mỗi tuple
ket_qua = sorted(list(cap_so), key=lambda x: x[0])

print("Danh sách các cặp số có tổng bằng", k, "là:", ket_qua)
```

### W6A11 [Tách số chẵn và lẻ từ tuple] Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt

```python
chuoi_nhap = input("Nhập tuple chứa các số nguyên (cách nhau bởi dấu cách): ")
tup_goc = tuple(int(x) for x in chuoi_nhap.split())

# Tách thành tuple chẵn và lẻ
tup_chan = tuple(x for x in tup_goc if x % 2 == 0)
tup_le = tuple(x for x in tup_goc if x % 2 != 0)

print("Tuple chẵn:", tup_chan)
print("Tuple lẻ:", tup_le)
```

### W6A12 [Tìm phần tử xuất hiện nhiều nhất] Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó

```python
chuoi_nhap = input("Nhập danh sách các số nguyên (cách nhau bởi dấu cách): ")
danh_sach = [int(x) for x in chuoi_nhap.split()]

# Đếm tần suất
dem = {}
for so in danh_sach:
    if so in dem:
        dem[so] += 1
    else:
        dem[so] = 1

# Tìm số lần xuất hiện cao nhất (max value)
max_lan = max(dem.values())

# Lọc ra danh sách các số đạt được số lần max đó
cac_so_max = []
for k, v in dem.items():
    if v == max_lan:
        cac_so_max.append(k)

print("Số xuất hiện nhiều nhất (nhỏ nhất nếu có nhiều):", min(cac_so_max))
```

### W6A13 [Đảo ngược key và value của dictionary] Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In ra dictionary mới.

```python
nhap = input("Nhập dictionary (key:value cách nhau khoảng trắng, VD: a:1 b:2 c:3): ")

dict_goc = {}
if nhap:
    for cap in nhap.split():
        k, v = cap.split(':')
        dict_goc[k] = v

dict_dao_nguoc = {}
for k, v in dict_goc.items():
    # Gán value của dict gốc làm key cho dict mới, và ngược lại
    dict_dao_nguoc[v] = k

print("Dictionary ban đầu:", dict_goc)
print("Dictionary đảo ngược:", dict_dao_nguoc)
```

### W6A14 [Giao của hai danh sách] Nhập vào hai dòng, mỗi dòng là một danh sách các số nguyên. Hãy tìm cácphần tử chung (giao) của hai danh sách này và in ra một list mới chứa các phần tử chung đó, không trùng lặp

```python
nhap1 = input("Nhập danh sách 1 (cách nhau bởi dấu cách): ")
nhap2 = input("Nhập danh sách 2 (cách nhau bởi dấu cách): ")

list1 = [int(x) for x in nhap1.split()]
list2 = [int(x) for x in nhap2.split()]

# Ép list về set và dùng toán tử & để lấy giao, sau đó ép ngược lại về list
giao = list(set(list1) & set(list2))

print("Các phần tử chung của hai danh sách là:", giao)
```

### W6A15 [Lọc dictionary theo giá trị] Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới.

```python
# Yêu cầu người dùng nhập dictionary, ví dụ: apple:50 banana:20 orange:80
chuoi_nhap = input("Nhập dictionary (các cặp key:value cách nhau khoảng trắng): ")
k = int(input("Nhập số nguyên k: "))

# Tạo dictionary ban đầu từ chuỗi nhập vào
dict_goc = {}
if chuoi_nhap:  # Kiểm tra xem người dùng có nhập gì không
    for cap in chuoi_nhap.split():
        key, value = cap.split(':')
        # Ép kiểu value về số nguyên để lát nữa so sánh với k
        dict_goc[key] = int(value)

# Tạo dictionary mới để lưu kết quả lọc
dict_moi = {}

# Duyệt qua từng cặp key và value trong dictionary gốc
for key, value in dict_goc.items():
    if value > k:
        dict_moi[key] = value  # Thêm vào dict mới nếu thỏa mãn điều kiện

print("Dictionary ban đầu:", dict_goc)
print(f"Dictionary mới (chứa các value > {k}):", dict_moi)
```
