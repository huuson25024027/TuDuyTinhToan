### Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân.

```python
a,b = map(int, input().split())
dien_tich_khu_dat = a * b
dien_tich_khu_vui_choi = 3.14 * (a/2)**2
dien_tich_trong_cay = dien_tich_khu_dat - dien_tich_khu_vui_choi
print(f"Diện tích khu vực trồng cây: {dien_tich_trong_cay:.2f}")
```

### Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)

```python
c = input()
if c.islower(): #Hàm .islower() kiểm tra xem tất cả các ký tự trong chuỗi có phải là chữ thường hay không.
    print(c.upper()) #Hàm .upper() chuyển đổi tất cả các ký tự trong chuỗi thành chữ hoa.
else:
    print(c.lower()) #Hàm .lower() chuyển đổi tất cả các ký tự trong chuỗi thành chữ thường.
```

### Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương ứng của A. Xét TH khi c = ‘A’ là đặc biệt

```python
ky_tu = input("Nhập chữ cái hoa: ")
# Xử lý trường hợp đặc biệt là chữ 'A'
if ky_tu == 'A':
    print("Chữ cái thường liền trước là: z")
    # Xử lý các chữ cái còn lại (từ B đến Z)
else:
# Chuyển chữ hoa thành chữ thường (ví dụ 'B' thành 'b')
    chu_thuong = ky_tu.lower()
# Biến chữ thành số, trừ đi 1, rồi biến lại thành chữ
    ma_so = ord(chu_thuong) - 1
    chu_truoc = chr(ma_so)
    print(f"Chữ cái thường liền trước là: {chu_truoc}")
```

### Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không, nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
a,b,c = map(int, input().split())
if a+b>c and a+c>b and b+c>a:
    print("3 cạnh này có thể tạo thành một tam giác")
    p = (a+b+c)/2
    dien_tich_tam_giac = (p*(p-a)*(p-b)*(p-c))**0.5
    print(f"Diện tích tam giác là: {dien_tich_tam_giac:.1f}")
else:
    print("3 cạnh này không thể tạo thành một tam giác")
```

- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng

```python
Ten_chu_ho = input("Nhập tên chủ hộ: ")
print(f"Ho va ten: {Ten_chu_ho}")
Chi_so_thang_truoc = int(input("Nhập chỉ số tháng trước: "))
Chi_so_thang_nay = int(input("Nhập chỉ số tháng này: "))
So_dien = Chi_so_thang_nay - Chi_so_thang_truoc
Tong_tien = 0
if So_dien <= 50:
    Tong_tien = So_dien * 1984
elif So_dien <= 100:
    Tong_tien = 50 * 1984 + (So_dien - 50) * 2050
elif So_dien <= 200:
    Tong_tien = 50 * 1984 + 50 * 2050 + (So_dien - 100) * 2380
elif So_dien <= 300:
    Tong_tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (So_dien - 200) * 2998
elif So_dien <= 400:
    Tong_tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (So_dien - 300) * 3350
else:
    Tong_tien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (So_dien - 400) * 3460
tien_tinh_thue = Tong_tien*1.08
print(f"Tien phai tra la: {int(tien_tinh_thue)} VND")
```

### 9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
so = int(input("Nhập số: "))
match so:
    case 0:
        print("Không")
    case 1:
        print("Một")
    case 2:
        print("Hai")
    case 3:
        print("Ba")
    case 4:
        print("Bốn")
    case 5:
        print("Năm")
    case 6:
        print("Sáu")
    case 7:
        print("Bảy")
    case 8:
        print("Tám")
    case 9:
        print("Chín")
    case _:
        print("Không hợp lệ")
```
