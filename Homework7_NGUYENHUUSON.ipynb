### W7A1 Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. Mẫu hàm: def binary_search(arr, left, right, target):


```python
def binary_search(arr, left, right, target):
    while left <= right:
        # Tìm vị trí ở giữa
        mid = (left + right) // 2
        
        # Nếu tìm thấy target ngay ở giữa
        if arr[mid] == target:
            return mid
        
        # Nếu target lớn hơn số ở giữa, ta bỏ qua nửa bên trái
        elif arr[mid] < target:
            left = mid + 1
            
        # Nếu target nhỏ hơn số ở giữa, ta bỏ qua nửa bên phải
        else:
            right = mid - 1
            
    # Nếu chạy hết vòng lặp mà không thấy target
    return -1

arr_test = [1, 2, 2, 3, 3, 4, 5, 6, 7, 8]
x = 9
result = binary_search(arr_test, 0, len(arr_test) - 1, x)
print(f"Chỉ số của {x} trong mảng là: {result}")
```

### W7A2 Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần trong danh sách. Mẫu hàm: def count_occurrences(arr, x) -> int:


```python

def count_occurrences(arr, x) -> int:
    count = 0
    for item in arr:
        if item == x:
            count += 1
    return count

arr_test = [1, 2, 2, 3, 3, 4, 5, 6, 7, 8]
target = 9
print(f"Số lần {target} xuất hiện: {count_occurrences(arr_test, target)} lần")
```

### W7A3 Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection Sort)


```python
#Bubble Sort
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        # n-i-1 vì sau mỗi vòng i, phần tử cuối đã nằm đúng chỗ
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                # Hoán đổi nếu số trước lớn hơn số sau
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr

arr_test = [3, 2, 4, 5, 7, 8, 6, 1, 2, 3]
print("Mảng ban đầu:", arr_test)
sorted_arr = bubble_sort(arr_test)
print("Mảng sau khi sắp xếp:", sorted_arr)
```

```python
#Selection Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        # Tìm vị trí của số nhỏ nhất trong phần còn lại
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Hoán đổi số nhỏ nhất tìm được với vị trí đầu tiên của đoạn
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr

arr_raw = [3, 2, 4, 5, 7, 8, 6, 1, 2, 3]
print(f"Mảng sau khi Bubble Sort: {bubble_sort(arr_raw.copy())}")
print(f"Mảng sau khi Selection Sort: {selection_sort(arr_raw.copy())}")
```

### W7A4 Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. Nếu có nhiều phần tử cùng số lần, trả về phần tử xuất hiện trước.


```python
def find_most_frequent(arr):
    if not arr:
        return None
        
    counts = {}
    for item in arr:
        if item in counts:
            counts[item] += 1
        else:
            counts[item] = 1

    most_freq_val = arr[0]
    max_count = counts[arr[0]]

    for item in arr:
        if counts[item] > max_count:
            max_count = counts[item]
            most_freq_val = item
    return f"{most_freq_val} xuat hien nhieu nhat, som nhat, {max_count} lan"

test_case = [2, 3, 4, 2, 2, 5]
print(find_most_frequent(test_case))
```

### W7A5 Cho danh sách các số nguyên và một số X, đếm tất cả các cặp (a, b) sao cho a+b=X.


```python
def count_pairs(arr, x):
    count = 0
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            # Kiểm tra tổng hai số
            if arr[i] + arr[j] == x:
                count += 1
                
    return count

arr_test = [1, 5, 7, -1, 5]
target = 6
print(f"Số lượng cặp có tổng bằng {target} là: {count_pairs(arr_test, target)}")
```

### W7A6 Bạn được cung cấp một danh sách các danh sách lồng nhau, hãy làm phẳng danh sách.


```python
def flatten_list(nested_list):
    flat = []     
    for sublist in nested_list:
        for item in sublist:
            flat.append(item)            
    return flat

danh_sach_long = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]
print(f"Danh sách sau khi làm phẳng: {flatten_list(danh_sach_long)}")
```

### W7A7 Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề).


```python
def length_of_LIS(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)

    for i in range(1, len(nums)):
        for j in range(i):
            if nums[i] > nums[j]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)

day_so = [10, 9, 2, 5, 3, 7, 101, 18]
print(f"Độ dài dãy con tăng nghiêm ngặt dài nhất là: {length_of_LIS(day_so)}")
```

### W7A8 Bạn được cung cấp một danh sách các khoảng thời gian interval = [[l1,r1], [l2, r2], ...] và một danh sách các truy vấn queries = [q1, q2, ...]. Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao cho li <= q <= ri. Nếu không tồn tại khoảng thời gian nào như vậy, hãy trả về -1 cho truy vấn đó. (Hint: binary search in the queries list)


```python
def min_interval_don_gian(intervals, queries):
    ket_qua = []

    for q in queries:
        ky_luc_ngan_nhat = float('inf') 
        for khoang in intervals:
            diem_dau = khoang[0]
            diem_cuoi = khoang[1]
            
            if diem_dau <= q <= diem_cuoi:
                do_dai = diem_cuoi - diem_dau + 1
                if do_dai < ky_luc_ngan_nhat:
                    ky_luc_ngan_nhat = do_dai

        if ky_luc_ngan_nhat == float('inf'):
            ket_qua.append(-1)  # Không có khoảng nào chứa q
        else:
            ket_qua.append(ky_luc_ngan_nhat) # Lưu lại kỷ lục ngắn nhất
            
    return ket_qua

intervals = [[1, 4], [2, 4], [3, 6], [4, 4]]
queries = [2, 3, 4, 5]
print("Output:", min_interval_don_gian(intervals, queries))
```

### W7A9 Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải nhỏ hơn nums[i]. Trả về danh sách các số đếm. (Efficient solution: using heap or binary search tree)


```python
def count_smaller_don_gian(nums):
    ket_qua = []

    for i in range(len(nums)):
        dem = 0  
        for j in range(i + 1, len(nums)):
            if nums[j] < nums[i]:
                dem += 1

        ket_qua.append(dem)       
    return ket_qua

day_so = [5, 2, 6, 1]
print(f"Số lượng phần tử nhỏ hơn bên phải của {day_so} là: {count_smaller_don_gian(day_so)}")

```

### W7A10 Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m. Tức là, với bất kỳ mảng con nums[i:j] nào, hãy tính sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể. Gợi ý: Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).


```python
def max_subarray_sum_mod_don_gian(nums, m):
    ky_luc_max = 0
    
    for i in range(len(nums)):
        tong_hien_tai = 0
        for j in range(i, len(nums)):
            tong_hien_tai += nums[j]
            phan_du = tong_hien_tai % m
            if phan_du > ky_luc_max:
                ky_luc_max = phan_du
                
    return ky_luc_max

day_so = [3, 3, 9, 9, 5]
modulo = 7
print(f"Tổng mảng con lớn nhất (modulo {modulo}) là: {max_subarray_sum_mod_don_gian(day_so, modulo)}")
```

### W7A12 Sắp xếp các cụm từ dưới đây theo thứ tự xuất hiện trong từ điển. Đầu vào: 4 cụm từ xuất hiện trên bốn dòng tương ứng với các phương án: A, B, C và D. Đầu ra: thứ tự đúng của các phương án, cách nhau bởi dấu cách.


```python
def sap_xep_phuong_an(A, B, C, D):
    danh_sach = [(A, 'A'), (B, 'B'), (C, 'C'), (D, 'D')]
    danh_sach.sort()
    ket_qua = []
    for item in danh_sach:
        ket_qua.append(item[1])
    return " ".join(ket_qua)
print(f"Thứ tự đúng: {sap_xep_phuong_an('May', 'Troi', 'Cua', 'Em')}") 
```

### W7A13 Dãy hài hòa là một dãy thỏa mãn điều kiện giá trị lớn nhất và giá trị nhỏ nhất của dãy khác nhau đúng bằng 1. Viết hàm trả về độ dài của dãy hài hòa con dài nhất có thể có trong mảng giá trị nguyên arr có n phần tử.


```python
def day_hai_hoa_dai_nhat(arr):
    dem = {}
    for so in arr:
        if so in dem:
            dem[so] += 1
        else:
            dem[so] = 1
            
    ky_luc = 0

    for so in dem:
        if (so + 1) in dem:
            do_dai = dem[so] + dem[so + 1]           
            if do_dai > ky_luc:
                ky_luc = do_dai                
    return ky_luc

day_so = [1, 3, 2, 2, 5, 2, 3, 7]
print(f"Độ dài dãy hài hòa con dài nhất: {day_hai_hoa_dai_nhat(day_so)}") 
```

### W7A14 Viết một chương trình tìm kiếm tất cả các số 7 có trong một dãy số. Đầu vào: số nguyên n trên một dòng và n số nguyên ở dòng kế tiếp. Đầu ra: Vị trí của các số 7 trong dãy cách nhau bởi dấu cách, vị trí lớn hơn đứng trước và bắt đầu tính từ 0. Nếu không tìm được số 7 nào in ra "Not found".


```python
def tim_vi_tri_so_7(arr):
    vi_tri = []
    for i in range(len(arr)):
        if arr[i] == 7:
            vi_tri.append(i)

    if len(vi_tri) == 0:
        return "Not found"
    
    vi_tri.sort(reverse=True)
    ket_qua = []
    for vt in vi_tri:
        ket_qua.append(str(vt))
    return " ".join(ket_qua)

day_so = [0, 4, 8, 2, 4, 8, 6, 1, 3, 2, 8, 5, 1, 6, 5, 4, 7, 7, 2, 0]
print(f"Vị trí các số 7: {tim_vi_tri_so_7(day_so)}") 
```

### W7A15 Nemo được mời tới tham dự hội nghị bàn tròn giữa các nhân vật hoạt hình. Tất cả các khách mời sẽ được gửi thiệp mời kèm theo một danh sách khách mời. Nemo muốn biết các nhân vật ngồi cạnh mình là ai để có thể chuẩn bị quà cho họ. Hãy giúp Nemo tìm tên của các nhân vật này, biết rằng tên trong danh sách khách mời được sắp xếp theo vị trí ngồi. Đầu vào: số nguyên n - số lượng khách mời trong danh sách trên một dòng, n dòng tiếp theo là tên của các vị khách mời. Đầu ra: Tên của hai người ngồi cạnh Nemo theo định dạng "<name1> and <name2>" (name1 là tên của người đứng trước và name2 là tên của người đứng sau Nemo trong danh sách).


```python
def tim_hang_xom_nemo(danh_sach_khach):
    n = len(danh_sach_khach)
    vi_tri_nemo = -1
    for i in range(n):
        if danh_sach_khach[i] == "Nemo":
            vi_tri_nemo = i
            break

    vi_tri_truoc = vi_tri_nemo - 1
    if vi_tri_truoc < 0:
        vi_tri_truoc = n - 1

    vi_tri_sau = vi_tri_nemo + 1
    if vi_tri_sau >= n:
        vi_tri_sau = 0
        
    nguoi_truoc = danh_sach_khach[vi_tri_truoc]
    nguoi_sau = danh_sach_khach[vi_tri_sau]
    
    return f"{nguoi_truoc} and {nguoi_sau}"

khach_moi = ["Anastatia", "Alice", "Ariel", "Aurora", "Nemo"]
print(f"Hai hàng xóm của Nemo là: {tim_hang_xom_nemo(khach_moi)}") 
```
