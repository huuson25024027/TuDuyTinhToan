### W2A3 Viết chương trình nhận vào hai số nguyên a, b từ bàn phím. In ra tông, hiệu, tích, phần nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập phân) theo định dạng đúng của output

````python
a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))
print("Tổng của hai số là:", a + b)
print("Hiệu của hai số là:", a - b)
print("Tích của hai số là:", a * b)
print("Phần nguyên của thương là:", a // b)
print("Phần dư của thương là:", a % b)
print(f"Thương của hai số là: {a / b:.2f}")
````

### W2A4 Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
* Các điểm a1, b1, c1 được nhân hệ số 1,
* Các điểm a2, b2 được nhân hệ số 2,
* Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10

----

**Đầu vào**: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được phân tách nhau bởi duy nhất 1 dấu cách.

**Đầu ra** : Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.

````python
a1, b1, c1, a2, b2, a3 = map(int, input("Nhập hệ số a1, b1, c1, a2, b2, a3: ").split())
diem_trung_binh = ((a1+b1+c1)+(a2+b2)*2+a3*3) / 10
print(f"Điểm trung bình là: {diem_trung_binh:.1f}")
````

### W2A8 Fahrenheit, hay độ F, là một thang nhiệt độ được đặt theo tên nhà vật lý người Đức Daniel Gabriel Fahrenheit (1686–1736). Dựa vào nhiệt độ Celcius được nhập từ bàn phím, hãy chuyển sang độ F theo công thức: F = 9/5*C + 32 và in ra màn hình kết quả. Kết quả làm tròn đến hai chữ số thập phân đằng sau dấu phẩy.

````python
C = float(input("Nhập số C: "))
F = (C * 9/5) + 32
print(f"{C} độ C tương đương với {F:.2f} độ F")
````
