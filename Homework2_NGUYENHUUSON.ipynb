### Bài 1: Nhập vào 2 cạnh của một hình chữ nhật. In ra màn hình diện tích và chu vi của nó.  

```python
a,b = map(int, input("Nhập vào 2 cạnh của hình chữ nhật: ").split())
dien_tich = a * b
print("Diện tích hình chữ nhật là:", dien_tich)
chu_vi = 2 * (a + b)
print("Chu vi hình chữ nhật là:", chu_vi)
```

### Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu  không thì in  ra màn  hình  'Khong la ba  canh cua mot tam  giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a,b,c = map(int, input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    elif a**2 + b**2 == c**2 or a**2 + c**2 == b**2 or b**2 + c**2 == a**2:
        print("Tam giác vuông") 
    else:
        print("Tam giác thường")
    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print("Diện tích tam giác là:", dien_tich)
    print("Chu vi tam giác là:", chu_vi)
else:
    print("Không phải là tam giác")
```

### Bài 4: Viết chương trình giải phương trình bậc 1 (ax + b = 0) 

```python
def ptb1(a, b):
    if a == 0:
        if b == 0:
            return "Vô số nghiệm"
        else:
            return "Vô nghiệm"
    else:
        x = -b / a
        return f"Nghiệm của phương trình là: {x}"
a, b = map(float, input("Nhập hệ số a và b của phương trình ax + b = 0: ").split())
result = ptb1(a, b)
print(result)
```

### Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
def ptb2(a, b, c):
    delta = b**2 - 4*a*c
    if delta > 0:
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        return f"Phương trình có hai nghiệm phân biệt: x1 = {x1}, x2 = {x2}"
    elif delta == 0:
        x = -b / (2*a)
        return f"Phương trình có nghiệm kép: x = {x}"
    else:
        return "Phương trình vô nghiệm"
a, b, c = map(float, input("Nhập hệ số a, b và c của phương trình ax^2 + bx + c = 0: ").split())
result = ptb2(a, b, c)
print(result)
```

### Bài 8: Giải hệ phương trình tuyến tính 
### ax + by = m
### cx + dy = n

```python
def giai_he_phuong_trinh(a, b, m, c, d, n):
    # Tính các định thức
    D = a*d - b*c
    Dx = m*d - b*n
    Dy = a*n - m*c  
    if D != 0:
        x = Dx / D
        y = Dy / D
        return f"Hệ có nghiệm duy nhất: x = {x}, y = {y}"
    else:
        if Dx == 0 and Dy == 0:
            return "Hệ có vô số nghiệm"
        else:
            return "Hệ vô nghiệm"
print("Giải hệ phương trình:")
print("ax + by = m")
print("cx + dy = n")
a, b, m = map(int, input("Nhập a, b, m (cách nhau khoảng trắng): ").split())
c, d, n = map(int, input("Nhập c, d, n (cách nhau khoảng trắng): ").split())
ket_qua = giai_he_phuong_trinh(a, b, m, c, d, n)
print(ket_qua)
```

### Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây.  

```python
x = int(input("Nhập vào thời gian: giây"))
gio = x // 3600
phut = (x % 3600) // 60
giay = x % 60
print(f"{gio} giờ {phut} phút {giay} giây")
```

### Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x,y = map(int, input("Nhập vào tọa độ điểm A (x y): ").split())
a,b = map(int, input("Nhập vào tọa độ tâm đường tròn (a b): ").split())
r = int(input("Nhập vào bán kính của đường tròn: "))
if r == ((x - a) ** 2 + (y - b) ** 2) ** 0.5:
    print(f"Điểm A({x}, {y}) nằm trên đường tròn tâm ({a}, {b}) bán kính {r}")
else:
    print(f"Điểm A({x}, {y}) không nằm trên đường tròn tâm ({a}, {b}) bán kính {r}")
```
