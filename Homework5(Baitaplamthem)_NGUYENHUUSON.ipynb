### W5A16 Viết hàm kiểm tra một số có phải số Amstrong hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_armstrong(x):
    if x < 0:
        return False
    
    chuoi_so = str(x)
    so_luong_chu_so = len(chuoi_so)
    tong = 0
    
    for chu_so in chuoi_so:
        tong += int(chu_so) ** so_luong_chu_so
        
    return tong == x

so_nguoi_dung_nhap = int(input("Mời bạn nhập vào một số nguyên: "))
ket_qua = is_armstrong(so_nguoi_dung_nhap)

# Kiểm tra kết quả trả về từ hàm và in ra màn hình
if ket_qua == True:
    print(f"Chính xác! {so_nguoi_dung_nhap} là số Armstrong.")
else:
    print(f"Rất tiếc, {so_nguoi_dung_nhap} không phải là số Armstrong.")
```

### W5A17 Viết hàm kiểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_perfect_square(x):
    if x < 0: return False
    can_bac_hai = int(x ** 0.5)
    return can_bac_hai * can_bac_hai == x

so_nhap = int(input("Nhập số nguyên x: "))
if is_perfect_square(so_nhap):
    print(f"{so_nhap} là số chính phương.")
else:
    print(f"{so_nhap} không phải là số chính phương.")
```

### W5A18 Viết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_palindrome(x):
    if x < 0: return False
    so_ban_dau, so_dao_nguoc = x, 0
    
    while x > 0:
        so_dao_nguoc = (so_dao_nguoc * 10) + (x % 10)
        x //= 10  # Phép chia lấy nguyên để cắt bỏ chữ số cuối
        
    return so_ban_dau == so_dao_nguoc

so_nhap = int(input("Nhập số nguyên x: "))
if is_palindrome(so_nhap):
    print(f"{so_nhap} là số đối xứng.")
else:
    print(f"{so_nhap} không phải số đối xứng.")
```

### W5A19 Viết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_magic_number(x):
    x = abs(x)
    while x > 9:  # Lặp đến khi tổng chỉ còn 1 chữ số
        tong_chu_so = 0
        while x > 0:
            tong_chu_so += x % 10
            x //= 10
        x = tong_chu_so
    return x == 1

so_nhap = int(input("Nhập số nguyên x: "))
if is_magic_number(so_nhap):
    print(f"{so_nhap} là số kỳ diệu.")
else:
    print(f"{so_nhap} không phải số kỳ diệu.")
```

### W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def detect_number_type(s):
    try:
        int(s)
        return 'int'
    except ValueError:
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'

chuoi_nhap = input("Nhập vào một chuỗi (s): ")
kieu_du_lieu = detect_number_type(chuoi_nhap)
print(f"Kiểu dữ liệu của chuỗi là: {kieu_du_lieu}")
```

### W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a, b, c):
    return (a == b * c) or (b == a * c) or (c == a * b)

a = int(input("Nhập số a: "))
b = int(input("Nhập số b: "))
c = int(input("Nhập số c: "))

if is_one_product_of_others(a, b, c):
    print("Đúng! Có một số bằng tích hai số còn lại.")
else:
    print("Sai! Không có số nào bằng tích hai số còn lại.")
```

### W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.

```python
def quadratic_roots_count(a, b, c):
    if a == 0:
        return 1 if b != 0 else 0 
        
    delta = b**2 - 4*a*c
    if delta > 0: return 2
    elif delta == 0: return 1
    else: return 0

print("Giải phương trình: ax^2 + bx + c = 0")
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

so_nghiem = quadratic_roots_count(a, b, c)
print(f"Phương trình có {so_nghiem} nghiệm thực.")
```

### W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không.

```python
def are_collinear(x1, y1, x2, y2, x3, y3):
    # Dùng tích chéo thay vì chia hệ số góc để tránh lỗi chia cho 0
    return (y2 - y1) * (x3 - x2) == (y3 - y2) * (x2 - x1)

print("Nhập tọa độ 3 điểm A, B, C:")
x1, y1 = float(input("x1: ")), float(input("y1: "))
x2, y2 = float(input("x2: ")), float(input("y2: "))
x3, y3 = float(input("x3: ")), float(input("y3: "))

if are_collinear(x1, y1, x2, y2, x3, y3):
    print("Ba điểm thẳng hàng.")
else:
    print("Ba điểm không thẳng hàng.")
```

### W5A25 Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),(0,3)], p = (2,1)









```python
def dien_tich_tam_giac(p1, p2, p3):
    return 0.5 * abs(p1[0]*(p2[1] - p3[1]) + p2[0]*(p3[1] - p1[1]) + p3[0]*(p1[1] - p2[1]))

def point_in_quadrilateral(quad, p):
    A, B, C, D = quad[0], quad[1], quad[2], quad[3]
    dien_tich_tu_giac = dien_tich_tam_giac(A, B, C) + dien_tich_tam_giac(A, C, D)
    dien_tich_4_tam_giac = (dien_tich_tam_giac(p, A, B) + dien_tich_tam_giac(p, B, C) + 
                            dien_tich_tam_giac(p, C, D) + dien_tich_tam_giac(p, D, A))
    # Dùng round để tránh sai số nhỏ của số thực (float) trong Python
    return round(dien_tich_tu_giac, 4) == round(dien_tich_4_tam_giac, 4)

print("NHẬP TỌA ĐỘ 4 ĐỈNH TỨ GIÁC")
x1 = float(input("Nhập x1: "))
y1 = float(input("Nhập y1: "))
x2 = float(input("Nhập x2: "))
y2 = float(input("Nhập y2: "))
x3 = float(input("Nhập x3: "))
y3 = float(input("Nhập y3: "))
x4 = float(input("Nhập x4: "))
y4 = float(input("Nhập y4: "))
quad = [(x1,y1), (x2,y2), (x3,y3), (x4,y4)]

print("NHẬP TỌA ĐỘ ĐIỂM P")
px = float(input("Nhập x của P: "))
py = float(input("Nhập y của P: "))
p = (px, py)

if point_in_quadrilateral(quad, p):
    print("=> Điểm P NẰM TRONG (hoặc trên biên) tứ giác.")
else:
    print("=> Điểm P NẰM NGOÀI tứ giác.")
```

### W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_nonzero_digits(lst_numbers):
    count = 0
    for num in lst_numbers:
        chuoi_so = str(abs(num))  # Dùng abs() để bỏ qua dấu trừ nếu là số âm
        for chu_so in chuoi_so:
            if chu_so != '0':
                count += 1
    return count

danh_sach_nhap = list(map(int, input("Nhập các số nguyên (cách nhau bởi dấu cách): ").split()))
print("Số lượng chữ số khác 0 là:", count_nonzero_digits(danh_sach_nhap))
```

### W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def calculate_angle_ABC(x1, y1, x2, y2, x3, y3):
    BA = (x1 - x2, y1 - y2)
    BC = (x3 - x2, y3 - y2)
    
    tich_vo_huong = (BA[0] * BC[0]) + (BA[1] * BC[1])
    do_dai_BA = math.sqrt(BA[0]**2 + BA[1]**2)
    do_dai_BC = math.sqrt(BC[0]**2 + BC[1]**2)
    
    cos_B = tich_vo_huong / (do_dai_BA * do_dai_BC)
    goc_radian = math.acos(cos_B)
    
    return math.degrees(goc_radian)

print("NHẬP TỌA ĐỘ 3 ĐIỂM")
x1 = float(input("Nhập x của A: "))
y1 = float(input("Nhập y của A: "))
x2 = float(input("Nhập x của B (đỉnh góc): "))
y2 = float(input("Nhập y của B (đỉnh góc): "))
x3 = float(input("Nhập x của C: "))
y3 = float(input("Nhập y của C: "))

goc = calculate_angle_ABC(x1, y1, x2, y2, x3, y3)
print(f"=> Góc ABC là: {goc} độ")
```

### W5A29 Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].

```python
def max_min_of_three(a, b, c):
    lon_nhat = max(a, b, c)
    nho_nhat = min(a, b, c)
    return [lon_nhat, nho_nhat]

a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
c = float(input("Nhập số c: "))

ket_qua = max_min_of_three(a, b, c)
print("=> Danh sách [max, min] là:", ket_qua)
```

### W5A30 Viết hàm tính hệ số nhị thức C(n, k).

```python
def giai_thua(x):
    kq = 1
    for i in range(1, x + 1):
        kq *= i
    return kq

def binomial_coefficient(n, k):
    if k < 0 or k > n:
        return 0
    return giai_thua(n) // (giai_thua(k) * giai_thua(n - k))

n = int(input("Nhập số nguyên dương n: "))
k = int(input("Nhập số nguyên dương k: "))
print(f"=> C({n}, {k}) = {binomial_coefficient(n, k)}")
```

### W5A33 Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa, giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    gia_mo_cua = 10000
    gia_km = 15000
    phi_cho_moi_phut = 500
    
    tong_tien = gia_mo_cua + (km * gia_km) + (wait_minutes * phi_cho_moi_phut)
    
    # Từ 22h đến 5h sáng hôm sau
    if start_hour >= 22 or start_hour <= 5:
        tong_tien *= 1.20 
        
    if km > 20:
        tong_tien *= 0.90
        
    return int(tong_tien)

# --- Phần chạy thử ---
loai_xe = input("Nhập loại xe (VD: 4_cho, 7_cho): ")
gio_bat_dau = int(input("Nhập giờ bắt đầu đi (0-23): "))
so_km = float(input("Nhập số quãng đường đi được (km): "))
thoi_gian_cho = int(input("Nhập số phút chờ: "))

tien_cuoc = smart_taxi_fare(loai_xe, gio_bat_dau, so_km, thoi_gian_cho)
print(f"=> Tổng tiền cước taxi ước tính là: {tien_cuoc} VNĐ")
```

### W5A34 Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau. Nếu n ≤ 0 trả về 0.

```python
def kiem_tra_chu_so_khac_nhau(x):
    chuoi_so = str(x)
    # Nếu x là 122, chuoi_so là '122', len là 3. set('122') sẽ là {'1', '2'}, len là 2.
    return len(set(chuoi_so)) == len(chuoi_so)

def sum_1_to_n(n):
    if n <= 0:
        return 0
        
    tong = 0
    for i in range(1, n + 1):
        if kiem_tra_chu_so_khac_nhau(i):
            tong += i
    return tong

# --- Phần chạy thử ---
n = int(input("Nhập số nguyên n: "))
print(f"=> Tổng các số thỏa mãn từ 1 đến {n} là: {sum_1_to_n(n)}")
```
