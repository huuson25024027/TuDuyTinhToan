### W6A16 Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n, m = map(int, input("Nhập số hàng n và số cột m (cách nhau bởi dấu cách): ").split())
ma_tran = []
print(f"Nhập dữ liệu cho ma trận {n}x{m}:")
for i in range(n):
    # Nhập từng hàng, cắt chuỗi và ép kiểu sang số nguyên
    hang = list(map(int, input(f"Nhập hàng thứ {i}: ").split()))
    ma_tran.append(hang)

# In ma trận với định dạng 4 vị trí căn phải
print("\nMa trận vừa nhập là:")
for hang in ma_tran:
    for so in hang:
        # :>4 nghĩa là dành ra 4 khoảng trống, đẩy số sang bên phải, end="" để các số trên cùng 1 hàng in cạnh nhau thay vì xuống dòng
        print(f"{so:>4}", end="") 
    print() # Xuống dòng khi in xong một hàng
```

### W6A17 Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng 

```python
n = int(input("Nhập kích thước ma trận vuông n x n: "))

ma_tran = []
print(f"Nhập dữ liệu cho ma trận vuông {n}x{n}:")
for i in range(n):
    hang = list(map(int, input(f"Nhập hàng thứ {i}: ").split()))
    ma_tran.append(hang)

cheo_chinh = []
cheo_phu = []

# Duyệt 1 vòng lặp để lấy cả 2 đường chéo
for i in range(n):
    cheo_chinh.append(ma_tran[i][i])          # Hàng i, Cột i
    cheo_phu.append(ma_tran[i][n - 1 - i])    # Hàng i, Cột n - 1 - i

print(f"\nĐường chéo chính: {cheo_chinh}")
print(f"Đường chéo phụ: {cheo_phu}")
```

### W6A18 Nhập vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách nhau dấu cách).Tính và in ra tổng các phần tử trên cột k (hàng k) 

```python
n, m, k = map(int, input("Nhập n, m, k (cách nhau bởi dấu cách): ").split())
ma_tran = []
print(f"Nhập dữ liệu cho ma trận {n}x{m}:")
for i in range(n):
    hang = list(map(int, input(f"Nhập hàng thứ {i}: ").split()))
    ma_tran.append(hang)

# Tính tổng hàng k
# Cần kiểm tra xem chỉ số k có hợp lệ (nằm trong số lượng hàng) không
if 0 <= k < n:
    tong_hang = 0
    for so in ma_tran[k]:
        tong_hang += so
    print(f"\nTổng các phần tử trên hàng {k} là: {tong_hang}")
else:
    print(f"\nKhông tồn tại hàng thứ {k} trong ma trận.")

# Tính tổng cột k
# Cần kiểm tra k có nằm trong số lượng cột không
if 0 <= k < m:
    tong_cot = 0
    # Để lấy cột, ta phải duyệt qua từng hàng i, và lấy phần tử ở vị trí k của hàng đó
    for i in range(n):
        tong_cot += ma_tran[i][k]
    print(f"Tổng các phần tử trên cột {k} là: {tong_cot}")
else:
    print(f"Không tồn tại cột thứ {k} trong ma trận.")
```


### W6A19 Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list 

```python
def tim_vi_tri_max(lst):
    if len(lst) == 0:
        return -1 

    vi_tri_max = 0
    gia_tri_max = lst[0]
    
    for i in range(1, len(lst)):
        if lst[i] > gia_tri_max:
            gia_tri_max = lst[i]
            vi_tri_max = i
            
    return vi_tri_max

danh_sach = list(map(int, input("Nhập danh sách các số nguyên: ").split()))
ket_qua = tim_vi_tri_max(danh_sach) 
print(f"List ban đầu: {danh_sach}")
print(f"Vị trí có giá trị lớn nhất là: {ket_qua}") 
```

### W6A20 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
def tim_vi_tri_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

danh_sach = list(map(int, input("Nhập danh sách các số nguyên: ").split()))    
so_k = int(input("Nhập số cần tìm: "))
print(f"Vị trí đầu tiên của {so_k} là: {tim_vi_tri_k(danh_sach, so_k)}")
```

### W6A21 Nhập một dòng gồm các chuỗi (tách bởi dấu cách).Tạo dictionary ánh xạ chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện 

```python
chuoi_nhap = input("Nhập một dòng các chuỗi (cách nhau bởi dấu cách): ")
danh_sach_chuoi = chuoi_nhap.split()

tu_dien_vi_tri = {}

for i in range(len(danh_sach_chuoi)):
    tu = danh_sach_chuoi[i]
    if tu not in tu_dien_vi_tri:
        tu_dien_vi_tri[tu] = [i]
    else:
        tu_dien_vi_tri[tu].append(i)

# Chuyển các list thành tuple
for tu in tu_dien_vi_tri:
    tu_dien_vi_tri[tu] = tuple(tu_dien_vi_tri[tu])

print("\nKết quả Dictionary:")
print(tu_dien_vi_tri)
```

### W6A23 Viết hàm xóa đi các số <x có trong mảng

### W6A24 Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 02324000900023249

```python
def day_khong_ve_dau(lst):
    ket_qua = []
    
    # Lấy tất cả các số 0
    for so in lst:
        if so == 0:
            ket_qua.append(so)
            
    # Lấy tất cả các số khác 0
    for so in lst:
        if so != 0:
            ket_qua.append(so)
            
    return ket_qua

danh_sach_bai_24 = [0, 2, 3, 2, 4, 0, 0, 0, 9]
print(f"Kết quả đẩy số 0 về đầu: {day_khong_ve_dau(danh_sach_bai_24)}")
```

### W6A25 Cho dãy chỉ gồm N giá trị từ 0-9 .Kiểm tra xem giá trị nào không có trong mảng

```python
def kiem_tra_so_thieu(lst):
    cac_so_thieu = []
    
    # Duyệt qua các số từ 0 đến 9
    for i in range(10):
        # Nếu số i không có mặt trong list đầu vào
        if i not in lst:
            cac_so_thieu.append(i)
            
    return cac_so_thieu

danh_sach_25 = list(map(int, input("Nhập một danh sách các số nguyên: ").split())) 
print(f"Các số từ 0-9 không có trong {danh_sach_25} là: {kiem_tra_so_thieu(danh_sach_25)}")
```

### W6A26 Tính tổng các phân tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối

```python
def tong_cuc_dai(lst):
    tong = 0
    
    # Nếu list có ít hơn 3 phần tử thì không thể có cực đại địa phương ở giữa
    if len(lst) < 3:
        return 0
        
    # Duyệt từ phần tử thứ 2 (index 1) đến phần tử áp chót
    for i in range(1, len(lst) - 1):
        # Kiểm tra điều kiện cực đại: lớn hơn phần tử trước VÀ phần tử sau
        if lst[i] > lst[i-1] and lst[i] > lst[i+1]:
            tong += lst[i]
            
    return tong

danh_sach_26 = list(map(int, input("Nhập một danh sách các số nguyên: ").split())) 
print(f"Tổng các cực đại địa phương của {danh_sach_26} là: {tong_cuc_dai(danh_sach_26)}")
```

### W6A27 Cho một dãy số nguyên, Tìm và in ra vị trí các số < x (x là giá trị cho trước). nếu không có in ra -1 

```python
def tim_vi_tri_nho_hon(lst, x):
    ket_qua = []

    for i in range(len(lst)):
        if lst[i] < x:
            ket_qua.append(i)

    if len(ket_qua) == 0:
        return -1
    else:
        return ket_qua

danh_sach = list(map(int, input("Nhập một danh sách các số nguyên: ").split())) 
gia_tri_x = int(input("Nhập giá trị x: "))
print(f"Vị trí các số < {gia_tri_x} là: {tim_vi_tri_nho_hon(danh_sach, gia_tri_x)}")
```

### W6A28 Cho một dãy các số nguyên, kiểm tra xem nó là dây tăng thực sự hay giảm thực sự, hay vô danh (không tăng, không giảm) 

```python
def phan_loai_day_so(lst):
    if len(lst) < 2:
        return "Vô danh"

    la_day_tang = True
    la_day_giam = True

    # Duyệt qua các cặp số liền kề
    for i in range(len(lst) - 1):
        if lst[i] >= lst[i+1]:
            la_day_tang = False
            
        if lst[i] <= lst[i+1]:
            la_day_giam = False

    if la_day_tang:
        return "Tăng thực sự"
    elif la_day_giam:
        return "Giảm thực sự"
    else:
        return "Vô danh"

danh_sach = list(map(int, input("Nhập một danh sách các số nguyên: ").split()))
print(f"Loại dãy số là: {phan_loai_day_so(danh_sach)}")
```
