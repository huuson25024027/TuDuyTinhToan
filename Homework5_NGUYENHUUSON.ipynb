### W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input("Nhập số cần kiểm tra (nguyên tố): "))
print(f"Kết quả: {is_prime(n)}")
```

### W5A4 Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def is_perfect(n: int) -> bool:
    if n < 2:
        return False
    tong = 0
    for i in range(1, n):
        if n % i == 0:
            tong += i
    return tong == n

n = int(input("Nhập số cần kiểm tra (hoàn hảo): "))
print(f"Kết quả: {is_perfect(n)}")
```

### W5A5 Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 

```python
def find_k(lst: list, k: int) -> int:
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1

# Nhập list từ bàn phím (các số cách nhau bằng dấu cách)
chuoi_nhap = input("Nhập danh sách các số nguyên (cách nhau bởi dấu cách): ")
lst = [int(x) for x in chuoi_nhap.split()]

k = int(input("Nhập số k cần tìm: "))
print(f"Vị trí của {k} trong list là: {find_k(lst, k)}")
```

### W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 

```python
def factorial(n: int) -> int:
    kq = 1
    for i in range(2, n + 1):
        kq *= i
    return kq

n = int(input("Nhập số nguyên dương n để tính giai thừa: "))
print(f"Kết quả {n}! = {factorial(n)}")
```

Biết các toán tử đơn giản thực hiện phép toán lần lượt là: Phép trừ \( - \) , Phép cộng \( + \) , Phép chia \( / \), Phép nhân \( * \). Kết quả làm tròn đến chữ số thập phân thứ hai.
```python
def calculator(num1: float, num2: float, operator: str):
    if operator == '+':
        kq = num1 + num2
    elif operator == '-':
        kq = num1 - num2
    elif operator == '*':
        kq = num1 * num2
    elif operator == '/':
        if num2 == 0:
            return "Lỗi chia cho 0"
        kq = num1 / num2
    else:
        return "Toán tử không hợp lệ"
        
    return round(kq, 2)

num1 = float(input("Nhập số thứ nhất: "))
num2 = float(input("Nhập số thứ hai: "))
operator = input("Nhập phép toán (+, -, *, /): ")

print(f"Kết quả: {calculator(num1, num2, operator)}")
```

### W5A8 Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. Cho hai số nguyên  x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này. 

```python
def hamming_distance(x: int, y: int) -> int:
    val = x ^ y
    count = 0
    while val > 0:
        if val % 2 == 1:
            count += 1
        val //= 2
    return count

x = int(input("Nhập số nguyên x: "))
y = int(input("Nhập số nguyên y: "))
print(f"Khoảng cách Hamming giữa {x} và {y} là: {hamming_distance(x, y)}")
```

### W5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 

```python
def sum_of_digits(n: int) -> int:
    tong = 0
    while n > 0:
        tong += n % 10
        n //= 10
    return tong

n = int(input("Nhập số nguyên dương n: "))
print(f"Tổng các chữ số của {n} là: {sum_of_digits(n)}")
```

### W5A10 Hai từ được gọi là đẳng cấu nếu các kí tự trong một từ có thể được ánh xạ để được từ còn lại. Ánh xạ một kí tự nghĩa là thay đổi toàn bộ các kí tự đó trong từ bởi một kí tự khác. Thứ tự của các kí tự được giữ không đổi. Hai kí tự khác nhau không thể ánh xạ đến cùng một kí tự, nhưng một kí tự có thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành 'z', 'b' thành 'b' và 'c' thành 'x'. Viết hàm  trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai từ đầu vào có cùng độ dài. 

```python
def is_isomorphic(a: str, b: str) -> bool:
    if len(a) != len(b):
        return False
    
    map_a_to_b = {}
    mapped_b = []
    
    for i in range(len(a)):
        char_a = a[i]
        char_b = b[i]
        
        if char_a in map_a_to_b:
            if map_a_to_b[char_a] != char_b:
                return False
        else:
            if char_b in mapped_b:
                return False
            map_a_to_b[char_a] = char_b
            mapped_b.append(char_b)
            
    return True

a = input("Nhập từ a: ")
b = input("Nhập từ b: ")
print(f"Hai từ có đẳng cấu không?: {is_isomorphic(a, b)}")
```

### W5A12 Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và 0 <= |r| < |divisor|.

```python
from typing import Tuple

def divide_with_remainder(dividend: int, divisor: int) -> Tuple[int, int]:
    q = dividend // divisor
    r = dividend % divisor
    return q, r

dividend = int(input("Nhập số bị chia (dividend): "))
divisor = int(input("Nhập số chia (divisor): "))
print(f"Kết quả bộ (thương, dư): {divide_with_remainder(dividend, divisor)}")
```

### W5A14 Viết hàm compute(a, b, op) thực hiện phép tính a (+|-|*|/) b tùy theo ký tự op.

```python
def compute(a: float, b: float, op: str):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return "Lỗi chia cho 0"
        return a / b
    return "Phép toán không hợp lệ"

a = float(input("Nhập số a: "))
b = float(input("Nhập số b: "))
op = input("Nhập phép toán (+, -, *, /): ")
print(f"Kết quả phép tính: {compute(a, b, op)}")
```

### W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông. 

```python
def is_right_triangle(a: float, b: float, c: float) -> bool:
    if a <= 0 or b <= 0 or c <= 0:
        return False
    
    if a*a + b*b == c*c:
        return True
    if a*a + c*c == b*b:
        return True
    if b*b + c*c == a*a:
        return True
        
    return False

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
print(f"Ba số này có phải 3 cạnh tam giác vuông không?: {is_right_triangle(a, b, c)}")
```
