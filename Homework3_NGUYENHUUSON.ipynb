### W3A1 Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ  NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = int(input("Nhập vào số nguyên dương n: "))
if 100 <= n <= 999:
    hang_tram = n // 100
    hang_chuc = (n % 100) // 10
    hang_don_vi = n % 10
    so_dao_nguoc = hang_don_vi * 100 + hang_chuc * 10 + hang_tram
    print("SỐ NGUYÊN đảo ngược của n là:", so_dao_nguoc)
else:
    print("Vui lòng nhập một số nguyên dương có ba chữ số.")
```

### W3A3 Kiểm tra xem một số có phải là lũy thừa của 2 hay không. Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không. Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input("Nhập vào số nguyên n: "))
if n>=0 and n&(n-1) == 0:
    print(n, "là lũy thừa của 2.")
else:
    print(n, "không phải là lũy thừa của 2.")
```

### W3A4 Viêt chương trình chia một số m cho n, kết quả thu được làm tròn xuống

```python

import math
m,n = map(int, input("Nhập hai số: ").split())
ket_qua = m / n
# Sử dụng hàm math.floor() để lấy số nguyên lớn nhất nhỏ hơn hoặc bằng kết quả
lam_tron_xuong = math.floor(ket_qua)
print(f"Kết quả làm tròn xuống: {lam_tron_xuong}")
```

### W3A5 Viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên 

```python
m,n = map(int, input("Nhập hai số: ").split())
x = m/n
if m%n == 0:
    print("Kết quả thu được là:", int(x))
else:
    print("Kết quả thu được là:", int(x+1))
```
