### W3A1 Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ  NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = int(input("Nhập vào số nguyên dương n: "))
if 100 <= n <= 999:
    hang_tram = n // 100
    hang_chuc = (n % 100) // 10
    hang_don_vi = n % 10
    so_dao_nguoc = hang_don_vi * 100 + hang_chuc * 10 + hang_tram
    print("SỐ NGUYÊN đảo ngược của n là:", so_dao_nguoc)
else:
    print("Vui lòng nhập một số nguyên dương có ba chữ số.")
```

### W3A11 Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều", "Tam giác cân", hay "Tam giác thường". 

```python
a,b,c = map(int, input("Nhập ba số: ").split())
if a+b>c and a+c>b and b+c>a:
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải là tam giác")
```
