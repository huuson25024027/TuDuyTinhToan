### W4A23 Nhập n. Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
def tim_k_vong_lap(n):
    k = 0
    S = 0  
    # Kiểm tra xem nếu cộng thêm (k + 1) thì tổng có nhỏ hơn n không
    while S + (k + 1) < n:
        k += 1
        S += k       
    return k, S

n = int(input("Nhập số nguyên n: "))
k, S = tim_k_vong_lap(n)

print(f"Giá trị k lớn nhất cần tìm là: {k}")
print(f"Khi đó S({k}) = {S} (nhỏ hơn {n})")
```

```python
import math
def tim_k_toan_hoc(n):
    # Tránh trường hợp n quá nhỏ (ví dụ n <= 1)
    if n <= 1:
        return 0, 0
        
    # Tính giá trị nghiệm dương của phương trình k^2 + k - 2n = 0
    k_float = (math.sqrt(1 + 8 * n) - 1) / 2  
    # Do cần S(k) "nhỏ hơn hẳn" n, ta xử lý bằng hàm ceil (làm tròn lên) rồi trừ 1
    k = math.ceil(k_float) - 1    
    # Tính lại tổng S
    S = k * (k + 1) // 2
    return k, S

n = int(input("Nhập số nguyên n: "))
k, S = tim_k_toan_hoc(n)

print(f"Giá trị k lớn nhất cần tìm là: {k}")
print(f"Khi đó S({k}) = {S} (nhỏ hơn {n})")
```

### W4A24 Nhập vào A, tìm n nhỏ nhất sao cho: 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
def tim_n_nho_nhat(A):
    n = 0
    S = 0.0 
    # Tiếp tục tăng n và cộng thêm 1/n cho đến khi S lớn hơn A
    while S <= A:
        n += 1
        S += 1 / n
        
    return n, S
A = float(input("Nhập giá trị A: "))
n, S = tim_n_nho_nhat(A)

print(f"Số n nhỏ nhất cần tìm là: {n}")
print(f"Khi đó tổng S = {S} (nhỏ nhất mà vẫn > {A})")
```

### W4A25 Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
so_lon_nhat = None
so_nho_nhat = None
print("Nhập vào các số nguyên (nhập -1 để dừng):")
while True:
    n = int(input("Nhập số: "))  
    # Kiểm tra điều kiện dừng ngay lập tức
    if n == -1:
        break       
    # Xử lý gán giá trị cho số đầu tiên được nhập
    if so_lon_nhat is None or so_nho_nhat is None:
        so_lon_nhat = n
        so_nho_nhat = n
    else:
        # So sánh và cập nhật lại lớn nhất / nhỏ nhất cho các số tiếp theo
        if n > so_lon_nhat:
            so_lon_nhat = n
        if n < so_nho_nhat:
            so_nho_nhat = n

# In kết quả sau khi vòng lặp kết thúc
if so_lon_nhat is not None:
    print(f"\nSố lớn nhất trong dãy là: {so_lon_nhat}")
    print(f"Số nhỏ nhất trong dãy là: {so_nho_nhat}")
else:
    # Xử lý trường hợp ngoại lệ: Người dùng nhập -1 ngay từ lần đầu tiên
    print("\nBạn chưa nhập số nào ngoài -1.")
```

### W4A26 Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
def tim_fibonacci_lon_nhat(A):
    # Dãy Fibonacci bắt đầu với 1, 1
    f1 = 1
    f2 = 1   
    # Nếu A < 1, không có số Fibonacci dương nào thỏa mãn
    if A < 1:
        return None       
    # Tiếp tục sinh số Fibonacci tiếp theo cho đến khi f2 vượt quá A
    while f2 <= A:
        # Cập nhật f1 thành f2, và f2 thành tổng của 2 số
        f1, f2 = f2, f1 + f2       
    # Khi vòng lặp kết thúc (f2 > A), f1 chính là số Fibonacci lớn nhất <= A
    return f1

A = int(input("Nhập số nguyên dương A: "))
ket_qua = tim_fibonacci_lon_nhat(A)
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {ket_qua}")
```

### W4A27 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và khoảng trắng). Dùng cách: for trên index; KHÔNG dùng hàm split(), len()

```python
chuoi = input("Nhập vào một chuỗi (chỉ gồm chữ và khoảng trắng): ")
# Tính chiều dài của chuỗi (vì không được dùng len)
chieu_dai = 0
for ky_tu in chuoi:
    chieu_dai += 1
# Đếm số từ bằng vòng lặp for trên index
so_tu = 0
for i in range(chieu_dai):
    # Nếu ký tự tại vị trí i là một chữ cái (không phải khoảng trắng)
    if chuoi[i] != ' ':
        # Nó sẽ là điểm bắt đầu của một từ nếu:
        # - Nó là ký tự đầu tiên của chuỗi (i == 0)
        # - HOẶC ký tự ngay trước nó (i - 1) là một khoảng trắng
        if i == 0 or chuoi[i-1] == ' ':
            so_tu += 1

print(f"Số từ trong chuỗi là: {so_tu}")
```

### W4A28 Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi. Dùng cách: for trên index; KHÔNG dùng hàm split() 

```python
chuoi = input("Nhập vào một chuỗi: ")
tu_dau_tien = ""
# Biến cờ để đánh dấu việc đã tìm thấy ký tự đầu tiên của từ hay chưa
da_bat_dau = False 
# Duyệt qua từng index của chuỗi
for i in range(len(chuoi)):
    if chuoi[i] != ' ':
        # Nếu gặp ký tự khác khoảng trắng, ghép nó vào từ và bật cờ
        tu_dau_tien += chuoi[i]
        da_bat_dau = True
    else:
        # Nếu gặp khoảng trắng VÀ cờ đã bật
        # Điều này có nghĩa là ta vừa đi hết từ đầu tiên và gặp khoảng trắng ngăn cách
        if da_bat_dau == True:
            break # Ngắt vòng lặp ngay lập tức vì đã lấy đủ từ đầu tiên

if tu_dau_tien != "":
    print(f"Từ đầu tiên trong chuỗi là: '{tu_dau_tien}'")
else:
    print("Chuỗi bạn nhập toàn khoảng trắng hoặc rỗng!")
```

### W4A29 Cho một chuỗi chỉ gồm các ký tự chữ cái thường và chữ cái hoa, hãy đổi ký tự hoa thành thường và thường thành hoa. 

```python
def dao_nguoc_hoa_thuong(chuoi):
    chuoi_ket_qua = ""
    
    for ky_tu in chuoi:
        # Kiểm tra xem ký tự có phải là chữ thường không
        if ky_tu.islower():
            chuoi_ket_qua += ky_tu.upper()
            
        # Kiểm tra xem ký tự có phải là chữ hoa không
        elif ky_tu.isupper():
            chuoi_ket_qua += ky_tu.lower()
            
        # Trong trường hợp chuỗi có lẫn ký tự khác (khoảng trắng, số), giữ nguyên
        else:
            chuoi_ket_qua += ky_tu
            
    return chuoi_ket_qua
chuoi_nhap = input("Nhập vào chuỗi cần đổi: ")
chuoi_moi = dao_nguoc_hoa_thuong(chuoi_nhap)

print(f"Chuỗi sau khi đổi: {chuoi_moi}")
```

### W4A31 Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số

```python
chuoi = input("Nhập vào một chuỗi: ")

dem_hoa = 0
dem_thuong = 0
dem_so = 0

# Duyệt qua từng ký tự trong chuỗi
for ky_tu in chuoi:
    if ky_tu.isupper():
        dem_hoa += 1
    elif ky_tu.islower():
        dem_thuong += 1
    elif ky_tu.isdigit():
        dem_so += 1
    # Bỏ qua các ký tự đặc biệt hoặc khoảng trắng (không rơi vào 3 trường hợp trên)

print(f"Số ký tự in hoa: {dem_hoa}")
print(f"Số ký tự in thường: {dem_thuong}")
print(f"Số ký tự số: {dem_so}")
```
