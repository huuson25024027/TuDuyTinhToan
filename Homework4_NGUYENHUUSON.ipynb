### W4A2 Viết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. Kiểm tra số này có phải là số nguyên tố hay không? 

```python
while True:
    n = int(input("Nhập vào một số nguyên dương: "))
    if n > 0:
        break  # Thoát khỏi vòng lặp nếu nhập đúng số > 0
    else:
        print("Số bạn nhập chưa đúng. Vui lòng nhập số lớn hơn 0!")
# Kiểm tra số nguyên tố
# Giả sử ban đầu n là số nguyên tố
la_so_nguyen_to = True
# Số nguyên tố phải lớn hơn 1
if n < 2:
    la_so_nguyen_to = False
else:
    # Thử chia n cho các số từ 2 đến n-1
    for i in range(2, n):
        if n % i == 0:
            la_so_nguyen_to = False # Nếu chia hết cho bất kỳ số nào thì không phải nguyên tố
            break 
if la_so_nguyen_to:
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```

### W4A6 Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn [a, b]. Mỗi dòng là một bộ test (a b)." 

```python
import math
def kiem_tra_nguyen_to(n):
    if n < 2:
        return False
    # Thuật toán tối ưu: Chỉ cần kiểm tra từ 2 đến căn bậc 2 của n
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
# Vòng lặp vô tận để đọc các bộ test liên tục
while True:
    try:
        # Đọc dữ liệu đầu vào. Nếu hết test, hệ thống tự ném ra lỗi EOFError
        line = input()  
        # Bỏ qua nếu lỡ đọc phải dòng trống
        if not line.strip():
            continue       
        # Tách chuỗi thành 2 số a và b
        a, b = map(int, line.split())     
        # Tính tổng các số nguyên tố trong đoạn [a, b]
        tong_nguyen_to = 0
        for so in range(a, b + 1):
            if kiem_tra_nguyen_to(so):
                tong_nguyen_to += so            
        # In kết quả của bộ test hiện tại
        print(tong_nguyen_to)       
    except EOFError:
        # Khi hệ thống báo đã hết file (hết bộ test), thoát khỏi vòng lặp
        break
```

### W4A7 Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n.

```python
n = int(input())
uoc_lon_nhat = 1
# Thử chia n cho các số từ i = 2
i = 2
# Chỉ cần lặp đến khi i * i <= n là đủ để tiết kiệm thời gian chạy
while i * i <= n:
    # Nếu n chia hết cho i, ta sẽ chia n cho i liên tục đến khi không chia được nữa
    while n % i == 0:
        uoc_lon_nhat = i  # Ghi nhận i là ước nguyên tố
        n //= i           # "Băm nhỏ" n    
    # Tăng i lên 1 để thử số tiếp theo
    i += 1
# Quan trọng: Nếu sau khi "băm nhỏ" xong vòng lặp trên mà n vẫn lớn hơn 1
# Điều đó chứng tỏ phần n còn sót lại này chính là một số nguyên tố, và nó cũng là ước nguyên tố lớn nhất luôn.
if n > 1:
    uoc_lon_nhat = n
print(uoc_lon_nhat)
```

### W4A8 Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
n = int(input())
buoc = 0
# Điều kiện vòng lặp: Chừng nào n chưa phải là số đối xứng thì còn lặp
while str(n) != str(n)[::-1]:
    # Tạo ra số đảo ngược bằng cách ép kiểu sang string, đảo ngược, rồi ép lại thành int
    so_dao_nguoc = int(str(n)[::-1]) 
    # Thực hiện phép toán n = n + reverse(n)
    n = n + so_dao_nguoc  
    # Tăng biến đếm số bước lên 1
    buoc += 1
print(f"{buoc} {n}")
```

### W4A9 Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu cách). Nếu không có, in ra (no number). 

```python
n = int(input())
ket_qua = []
# Biến đếm i để tạo số chính phương (i * i)
i = 0
# Duyệt các số chính phương <= n
while i * i <= n:
    so_chinh_phuong = i * i
    chuoi_so = str(so_chinh_phuong)   
    # Ép chuỗi vào set() để lọc các ký tự trùng nhau
    # Nếu độ dài của set bằng độ dài chuỗi ban đầu -> các chữ số đôi một khác nhau
    if len(set(chuoi_so)) == len(chuoi_so):
        ket_qua.append(chuoi_so)       
    i += 1
if ket_qua:
    # Nối các phần tử trong danh sách bằng dấu cách
    print(" ".join(ket_qua))
else:
    print("(no number)")
```

### W4A10 Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.

```python
n = int(input())
memo = {1: 1} # Lưu sẵn độ dài Collatz của 1 là 1 để làm điểm dừng
max_len = 0
best_x = 1
for i in range(1, n + 1):
    k = i
    count = 0
    while k not in memo: # Thuật toán rẽ nhánh Collatz, lặp đến khi gặp số đã được tính trước đó
        if k % 2 == 0:
            k = k // 2
        else:
            k = k * 3 + 1
        count += 1
    memo[i] = count + memo[k] # Cộng dồn độ dài mới với độ dài đã lưu trong memo
    if memo[i] > max_len: # Chỉ lấy lớn hơn hẳn (>) để đảm bảo nếu trùng độ dài thì ưu tiên số x nhỏ hơn xuất hiện trước
        max_len = memo[i]
        best_x = i
print(f"{best_x} {max_len}")
```

### W4A11 Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6

```python
n = int(input())
count = 0
if n % 2 == 0: # Bỏ qua ngay từ đầu nếu n là số lẻ
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            if i % 2 == 0: count += 1
            if (n // i) % 2 == 0 and i != n // i: count += 1 # Xét luôn ước đối xứng n//i và tránh đếm trùng nếu n là số chính phương
print(count)
```

### W4A13 Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật: Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại. Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
def tinh_tong_uoc(n):
    tong = 0
    for i in range(1, n): # Duyệt từ 1 đến n-1 (trừ chính nó)
        if n % i == 0:
            tong += i
    return tong
data = input().split()
a, b = int(data[0]), int(data[1])
# Kiểm tra tổng ước số này bằng số kia và ngược lại
if tinh_tong_uoc(a) == b and tinh_tong_uoc(b) == a:
    print("true")
else:
    print("false")
```

### W4A14 Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình tìm ước chung lớn nhất của hai số m và n.

```python
data = input().split()
m, n = int(data[0]), int(data[1])
# Thuật toán Euclid để tìm UCLN
while n != 0:
    m, n = n, m % n # Gán m bằng n, và n bằng số dư của m chia n
print(m)
```

### W4A15 "Vừa gà vừa chó, Bó lại cho tròn, Ba mươi sáu con, Một trăm chân chẵn" Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và số lượng chó. Ngược lại, in ra "invalid". 

```python
data = input().split()
tong_con, tong_chan = int(data[0]), int(data[1])
found = False
for ga in range(tong_con + 1):
    cho = tong_con - ga
    if (ga * 2 + cho * 4) == tong_chan: # Kiểm tra điều kiện tổng số chân
        print(f"{ga} {cho}")
        found = True
        break
if not found: print("invalid") # In invalid nếu không có số lượng gà, chó nào phù hợp
```
