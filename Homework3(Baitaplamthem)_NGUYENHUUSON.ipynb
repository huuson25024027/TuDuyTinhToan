### Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("Nhập chiều cao h: "))
for i in range(1, h + 1):
    if i == 1:
        # Dòng đầu tiên (đỉnh tam giác): In ra (h-1) khoảng trắng và 1 dấu *
        print(" " * (h - i) + "*")
    elif i == h:
        # Dòng cuối cùng (đáy tam giác): In ra (2*h - 1) dấu *
        print("*" * (2 * h - 1))
    else:
        # Các dòng ở giữa (cạnh bên): 
        # Cần khoảng trắng bên ngoài, 1 dấu *, khoảng trắng bên trong, và 1 dấu *
        khoang_trang_ngoai = " " * (h - i)
        khoang_trang_trong = " " * (2 * i - 3)
        print(khoang_trang_ngoai + "*" + khoang_trang_trong + "*")
```

### Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n 

```python
n = int(input("Nhập số nguyên dương n: "))
S = 0
ket_qua = 0
# Cho k chạy từ 1 đến n
for k in range(1, n):
    # Nhìn thử xem nếu cộng thêm k thì có vượt n không
    if S + k < n:
        S += k
        ket_qua = k  # Lưu lại giá trị k thỏa mãn
    else:
        # Nếu cộng thêm k mà lớn hơn hoặc bằng n thì dừng vòng lặp ngay
        break 
print(f"Số k lớn nhất cần tìm là: {ket_qua}")
print(f"Kiểm tra lại: S({ket_qua}) = {S} (nhỏ hơn {n})")
```

```python
n = int(input("Nhập số nguyên dương n: "))

k = 0
S = 0

# Vòng lặp sẽ tiếp tục chạy chừng nào việc cộng thêm số kế tiếp (k + 1) vẫn cho ra kết quả nhỏ hơn n
while S + (k + 1) < n:
    k += 1          # Tăng k lên 1 đơn vị
    S += k          # Cộng dồn k vào S (tương đương S = S + k)

print(f"Số k lớn nhất cần tìm là: {k}")
print(f"Kiểm tra lại: S({k}) = {S} (nhỏ hơn {n})")
```

### Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho: 1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A

```python
A = int(input("Nhập vào số nguyên dương A: "))
n = 0
S = 0.0 # Dùng 0.0 để biểu diễn số thực (vì chúng ta cộng phân số)
# Lặp chừng nào S còn nhỏ hơn hoặc bằng A
while S <= A:
    n += 1        # Tăng mẫu số lên 1
    S += 1 / n    # Cộng phân số 1/n vào tổng S
print(f" Với A = {A} n nhỏ nhất cần tìm là: {n}")
print(f"Kiểm tra lại: Tổng S = {S} (đã lớn hơn {A})")
```

### Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
# Khởi tạo biến với giá trị None (nghĩa là "chưa có gì")
max_val = None
min_val = None
while True:
    n = int(input("Nhập một số nguyên (nhập -1 để dừng): "))  
    # Kiểm tra ngay lập tức để không đưa số -1 vào tính toán
    if n == -1:
        break       
    # Trường hợp đây là số ĐẦU TIÊN được nhập
    if max_val is None and min_val is None:
        max_val = n
        min_val = n
    else:
        # Từ số thứ 2 trở đi, đem so sánh với kỷ lục hiện tại
        if n > max_val:
            max_val = n
        if n < min_val:
            min_val = n
# Xử lý trường hợp người dùng gõ -1 ngay từ lần đầu tiên (chưa có số nào)
if max_val is not None:
    print(f"Số lớn nhất là: {max_val}")
    print(f"Số nhỏ nhất là: {min_val}")
else:
    print("Bạn chưa nhập số nào!")
```

### Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input("Nhập số nguyên dương n: "))
temp = n
dem_chan = 0
dem_le = 0
while temp > 0:
    # Lấy ra chữ số cuối cùng
    chu_so = temp % 10    
    # Kiểm tra chẵn lẻ
    if chu_so % 2 == 0:
        dem_chan += 1
    else:
        dem_le += 1       
    # Cắt bỏ chữ số cuối cùng đi để vòng sau xét số tiếp theo
    temp //= 10  
print(f"Số {n} có {dem_chan} chữ số chẵn và {dem_le} chữ số lẻ.")
```

```python
# Nhận thẳng đầu vào dạng chuỗi 
chuoi_n = input("Nhập số nguyên dương n: ") 
dem_chan = 0
dem_le = 0
# Duyệt qua từng ký tự trong chuỗi
for ky_tu in chuoi_n:
    # Ép ký tự đó thành số để kiểm tra chẵn lẻ
    if int(ky_tu) % 2 == 0:
        dem_chan += 1
    else:
        dem_le += 1
print(f"Số {chuoi_n} có {dem_chan} chữ số chẵn và {dem_le} chữ số lẻ.")
```

### Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
def tong_chu_so_toan_hoc(n):
    tong = 0
    while n > 0:
        chu_so_cuoi = n % 10  # Lấy chữ số cuối cùng 
        tong = tong + chu_so_cuoi # Cộng dồn vào tổng
        n = n // 10     # Bỏ chữ số cuối cùng đi 
    return tong 
n = int(input("Nhập số nguyên dương n: "))
if n > 0:
    print(f"Tổng các chữ số là: {tong_chu_so_toan_hoc(n)}")
else:
    print("Vui lòng nhập số nguyên dương!")
```

### Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không 

```python
def la_luy_thua_cua_3_loop(n):
    # Số nguyên dương dạng 3^k luôn lớn hơn 0
    if n <= 0:
        return False  
    while n % 3 == 0:
        n = n // 3       
    # Nếu sau khi chia hết các số 3 mà n về đúng số 1 thì hợp lệ
    return n == 1
n = int(input("Nhập số nguyên dương n: "))
if la_luy_thua_cua_3_loop(n):
    print(f"{n} là số có dạng 3^k")
else:
    print(f"{n} không phải là số có dạng 3^k")
```

### Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
def tim_fibo_lon_nhat(A):
    # Khởi tạo 2 phần tử đầu tiên của dãy
    a = 1
    b = 1  
    # Chừng nào số tiếp theo (b) chưa vượt quá A thì tiếp tục chạy
    while b <= A:
        a, b = b, a + b       
    # Khi b > A, vòng lặp dừng. Giá trị thoả mãn điều kiện nằm ở a.
    return a
try:
    A = int(input("Nhập số nguyên dương A: "))
    if A > 0:
        ket_qua = tim_fibo_lon_nhat(A)
        print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {ket_qua}")
    else:
        print("Vui lòng nhập một số nguyên dương (lớn hơn 0).")
except ValueError:
    print("Vui lòng nhập một số nguyên hợp lệ!")
```

### Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
def dem_nguyen_to_chuoi(n):
    # Tập hợp các ký tự số nguyên tố
    tap_nguyen_to = {'2', '3', '5', '7'}
    dem = 0  
    # Duyệt từng chữ số trong chuỗi n
    for chu_so in str(n):
        if chu_so in tap_nguyen_to:
            dem += 1           
    return dem
# Nhập và chạy thử
n = int(input("Nhập số nguyên dương n: "))
if n > 0:
    print(f"Số lượng chữ số nguyên tố là: {dem_nguyen_to_chuoi(n)}")
else:
    print("Vui lòng nhập số nguyên dương!")
```
